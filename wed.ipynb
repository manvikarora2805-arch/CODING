```python
num = int(input("Enter a number: "))
if num > 1:
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            print(f"{num} is not a prime number.")
            break
    else:
        print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

```python
class polymorphosis:
    def display(self):
        print("m")
    def play(self):
        print("cricket")

class polymorphosisss:
    def display(self):
        print("d")
    def play(self):
        print("football")


d=polymorphosisss()
m=polymorphosis()
for i in (d,m):
    i.display()
    i.play()

```
